```
import torch

# mapping: src [0, 0, 1] , tgt [0, 1, 2, 2] (len 2, sum 3)

# src가 i -> 10,10,10 (2, 1, 3) -> 각각 tgt id에 통과시켜서 (2, 1, 2)가 됨,
# 즉 (0-2, 1-1, 2-2)로 sum 3이 충족된 상황,
# 왼쪽 값들을 src id에도 통과시키면 (0-2, 0-1, 1-2)가 되고 .. 0이 두개로 매핑되는건 상관없어서 종료 ?
# 이게 더 쉬운거 아녀 ??

# tgt가 i (dim=0)-> 7, 10, 10, 10  (2, 1, 0, 2) -> src id 통과 -> (1,0,0,1)
# 즉 (1-0, 0-1, 0-2, 1-3)로 len이 충족된 상황,
# 오른쪽 값들을 tgt id에도 통과시키면 (1-0, 0-1, 0-2, 1-2)가 되고 2와 매칭되는 두 케이스가 문제가 됨
# -> p를 직접 참조해서 더 높은 쪽 선택 (ex, 1-2) -> 최종 '1-0, 1-2, 0-1'로 [2,1]가 fertility


# ---- 위 src 기준 방식 사용


A = torch.tensor([[1, 2, 10, 0],
                  [4, 10, 5, 0],
                  [7, 8, 9, 10]])
src_map = torch.tensor([0,0,1])
tgt_map = torch.tensor([0,1,2,2])

argmax_indices = torch.argmax(A, dim=1)
sum_ok = tgt_map[argmax_indices]
fertility = []
ft= ''
for i, x in enumerate(sum_ok):
  ft+= (' ' + str(src_map[i].item()) + '-' + str(x.item()))
fertility.append(ft)
print(ft)


# 근데 또 생각하지못한.. 문제가.. src_subword 수 기준으로 후보들을 뽑았더니
# sent_tgt < src_subword 일 때 문장이 잘림 ..ㅋ  ...ㅋㅋ
```

## fertility label

```

#약 1시간 정도 (cpu의 16배)

import torch
import numpy as np
import time
import itertools

# 1차: de_37 and en_37 (37000 bpe version)
# 2차: de and en (raw dataset version)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

ft = []
ft_label = []
idx = 0
bef = time.time()

for src, tgt in zip(de_37, en_37):
    # if idx == 10:
    #   break


    if idx % 1000 == 0:
        now = time.time()
        print(f"{idx} done, took {np.round((now - bef) / 60, 3)} minutes")
        bef = now

    # pre-processing
    sent_src, sent_tgt = src.strip().split(), tgt.strip().split()
    token_src, token_tgt = [tokenizer.tokenize(word) for word in sent_src], [tokenizer.tokenize(word) for word in sent_tgt]
    wid_src, wid_tgt = [tokenizer.convert_tokens_to_ids(x) for x in token_src], [tokenizer.convert_tokens_to_ids(x) for x in token_tgt]
    ids_src, ids_tgt = tokenizer.prepare_for_model(list(itertools.chain(*wid_src)), return_tensors='pt', model_max_length=tokenizer.model_max_length, truncation=True)['input_ids'], tokenizer.prepare_for_model(list(itertools.chain(*wid_tgt)), return_tensors='pt', truncation=True, model_max_length=tokenizer.model_max_length)['input_ids']
    sub2word_map_src = []
    for i, word_list in enumerate(token_src):
        sub2word_map_src += [i for x in word_list]
    sub2word_map_tgt = []
    for i, word_list in enumerate(token_tgt):
        sub2word_map_tgt += [i for x in word_list]


    align_layer = 8
    threshold = 1e-3

    with torch.no_grad():
        ids_src = ids_src.to(device)
        ids_tgt = ids_tgt.to(device)

        out_src = model(ids_src.unsqueeze(0), output_hidden_states=True)[2][align_layer][0, 1:-1]
        out_tgt = model(ids_tgt.unsqueeze(0), output_hidden_states=True)[2][align_layer][0, 1:-1]

        dot_prod = torch.matmul(out_src, out_tgt.transpose(-1, -2)).to(device)
        softmax_tgtsrc = torch.nn.Softmax(dim=-2)(dot_prod)


        argmax_i = list(torch.argmax(softmax_tgtsrc, dim=1)) # src _ subword 수만큼
        argmax_j = list(torch.argmax(softmax_tgtsrc, dim=0)) # tgt _ subword 수만큼


    candid= []
    p=[]

    for src_sub_i, tgt_sub_i in enumerate(argmax_i):
      src_i = sub2word_map_src[src_sub_i]
      tgt_i = sub2word_map_tgt[tgt_sub_i]
      p.append(softmax_tgtsrc[src_sub_i, tgt_sub_i].item())
      candid.append( (src_i, tgt_i) )

    for tgt_sub_j, src_sub_j in enumerate(argmax_j):
      src_j = sub2word_map_src[src_sub_j]
      tgt_j = sub2word_map_tgt[tgt_sub_j]
      p.append(softmax_tgtsrc[src_sub_j, tgt_sub_j].item())
      candid.append( (src_j, tgt_j)  )

    candid = list(set(candid)) #이론적으로 (?) 항상 len(sent_tgt), len(sent_src)보다 큼
    # print(len(candid))

    combined = list(zip(p, candid))
    sorted_combined = sorted(combined, key=lambda pair: pair[0], reverse=True)
    _, sorted_candid = zip(*sorted_combined)

    # print(sorted_candid)

    ft_str= ''
    ft_label_temp = [0] * len(sent_src)


    # 결국 최소 sent_tgt 수 이상의 후보들 중에 sent_tgt 수만큼 선택되기만 하면 됨.
    for i, x in sorted_candid[:len(sent_tgt)]:
      ft_label_temp[i]+=1
      ft_str+= (' ' + str(i) + '-' + str(x))

    # print(ft_label_temp)
    # print(ft_str)

    # print(sum(ft_label_temp), len(sent_tgt))
    # print(len(ft_label_temp), len(sent_src))

    ft.append(ft_str)
    ft_label.append(ft_label_temp)
    idx += 1
```

```
import pickle

with open('ft_awsalign_37.pkl', 'wb') as file:
    pickle.dump(ft, file)

with open('ft_awsalign_37.pkl', 'wb') as file:
    pickle.dump(ft_label, file)

# with open('ftlabel_awsalign_37.pkl', 'wb') as file:
#     pickle.dump(ft, file)
```
